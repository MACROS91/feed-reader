```python
# Import models from Django
from fourchan.models import FourChan
from reddit.models import Reddit
```

```python
# Iterate through reddit posts and get contract addresses and TGs
posts = Reddit.objects.all()
#posts = FourChan.objects.all()

for post in posts:
    # Get addresses that match ETH address format
    contract_address = re.search(r'0x[a-fA-F0-9]{40}', post.selftext)
    if contract_address:
        print(contract_address.group(0))
    
    # Get TG url if in post
    tg_url = re.search('https://t.me(\/.*)?', post.selftext)
    if tg_url:
        tg_url = tg_url.group(0)
        tg_url = tg_url.replace('\\','')
        tg_url = tg_url.split(']')[0]
        tg_url = tg_url.split(')')[0]
        print(tg_url)
```
